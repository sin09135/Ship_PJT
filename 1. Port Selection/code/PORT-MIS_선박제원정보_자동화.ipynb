```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
import time
from selenium.webdriver.common.action_chains import ActionChains
import pyperclip
import pandas as pd
from selenium.common.exceptions import NoSuchElementException
```

```python

# Chrome WebDriver 설정
CHROME_DRIVER_PATH = 'chromedriver.exe'
service = Service(executable_path=CHROME_DRIVER_PATH)
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)

# 상세정보 xpath
상세정보_xpath_list = [
    '//*[@id="mf_tacMain_contents_M0177_body_grtg"]',
    '//*[@id="mf_tacMain_contents_M0177_body_cmdtyWtTon"]',
    '//*[@id="mf_tacMain_contents_M0177_body_vsslTotLt"]',
    '//*[@id="mf_tacMain_contents_M0177_body_shdth"]',
    '//*[@id="mf_tacMain_contents_M0177_body_vsslDrft"]',
    '//*[@id="mf_tacMain_contents_M0177_body_vsslDp"]',
    '//*[@id="mf_tacMain_contents_M0177_body_vsslLt"]',
    '//*[@id="mf_tacMain_contents_M0177_body_vsslNltyCd"]',
    '//*[@id="mf_tacMain_contents_M0177_body_vsslNltyNm"]'
]

############################################################### 시작 위치
start_index = 4565

# 웹 페이지 열기
driver.get('https://new.portmis.go.kr/portmis/websquare/websquare.jsp?w2xPath=/portmis/w2/main/index.xml&page=/portmis/w2/co/como/intro/nologin/UI-PM-CO-110.xml&menuId=0463&menuCd=M2144&menuNm=%EC%BB%A8%ED%85%90%EC%B8%A0%EC%A1%B0%ED%9A%8C')

# 시간 대기
time.sleep(3)

# 선박제원정보 들어가기
driver.find_element(By.XPATH, '//*[@id="mf_tacMain_contents_M2144_body_genInfoMenu_0_genInfoItem"]').click()

# 반복문으로 호출 코드 입력
for 호출부호 in new_df['호출부호'][start_index:]:
    try:
        # 호출코드입력창 클릭
        elem = driver.find_element(By.XPATH, '//*[@id="mf_tacMain_contents_M0177_body_srchVsslId"]')
        elem.click()

        # 호출 코드 지우기
        elem.send_keys(Keys.BACKSPACE * len(elem.get_attribute('value')))

        # 호출 코드 입력
        elem.send_keys(호출부호)

        # 검색 버튼 클릭
        driver.find_element(By.XPATH, '//*[@id="mf_tacMain_contents_M0177_body_wq_uuid_739_btnSearch"]/a').click()
        time.sleep(1.5)

        # 선박 제원 확인을 위한 표 클릭
        driver.find_element(By.XPATH, '//*[@id="mf_tacMain_contents_M0177_body_vsslListGrid_cell_0_1"]').click()
        time.sleep(2)

        # 데이터프레임에 값 채우기
        for idx, xpath in enumerate(상세정보_xpath_list):
            # ActionChains를 사용하여 엘리먼트를 더블클릭
            action = ActionChains(driver)
            action.double_click(driver.find_element(By.XPATH, xpath)).perform()

            # JavaScript를 사용하여 텍스트를 복사
            element_text = driver.execute_script("return window.getSelection().toString()")

            # 결과 출력
            print(f"{xpath}: {element_text}")

            # 데이터프레임에 값 넣기
            new_df.loc[new_df['호출부호'] == 호출부호, new_df.columns[idx + 1]] = element_text  # 호출부호 열은 0번째 열이므로, 열의 인덱스를 조정

        # 현재 처리한 호출부호의 인덱스 저장
        current_index = new_df[new_df['호출부호'] == 호출부호].index[0]

        # 다음에 다시 시작할 위치 저장
        start_index = current_index + 1

        # 잠시 대기
        time.sleep(2)

    except Exception as e:
        print(f"예외 발생: {e}")
        print("경고창 버튼 클릭 후 코드를 다시 시작합니다.")
        
        # 페이지 새로고침
        driver.refresh()
        time.sleep(5)
        driver.find_element(By.XPATH, '//*[@id="mf_tacMain_contents_M2144_body_genInfoMenu_0_genInfoItem"]').click()
        time.sleep(2)
        
        # 다음 호출부호로 진행
        start_index += 1
        continue
```
