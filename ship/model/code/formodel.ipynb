### 입항일시
- 입항일시를 ETA로 파악

```python
df = temp.copy()
```

```python
df['입항일시'] = pd.to_datetime(df['입항일시'])
df['ETA_Year'] = df['입항일시'].dt.year
df['ETA_Month'] = df['입항일시'].dt.month
df['ETA_Day'] = df['입항일시'].dt.day
df['ETA_Hour'] = df['입항일시'].dt.hour
df['ETA_Minute'] = df['입항일시'].dt.minute
```

```python
df.columns
```

## 해양기상데이터 merge

```python
wt20 = pd.read_csv(directory + '2020(울산해양기상).csv')
wt21= pd.read_csv(directory + '2021(울산해양기상).csv')
wt22 = pd.read_csv(directory + '2022(울산해양기상).csv')
```

```python
wt = pd.concat([wt20, wt21, wt22], ignore_index=True)
```

```python
wt.info()
```

```python
wt.drop(columns = '지점', inplace = True)
```

```python
wt
```

```python
df = df[df['ETA_Year'].isin([2020, 2021, 2022])] 
```

```python
df
```

```python
wt['일시'] = pd.to_datetime(wt['일시'])
wt['ETA_Year'] = wt['일시'].dt.year
wt['ETA_Month'] = wt['일시'].dt.month
wt['ETA_Day'] = wt['일시'].dt.day
wt['ETA_Hour'] = wt['일시'].dt.hour
wt['ETA_Minute'] = wt['일시'].dt.minute
```

```python
wt[['ETA_Year','ETA_Month', 'ETA_Day', 'ETA_Hour','ETA_Minute']]
```

```python
wt
```

```python
merged = pd.merge(df, wt, on = ['ETA_Year','ETA_Month', 'ETA_Day', 'ETA_Hour'], how = 'left')
```

```python
merged.info()
```

```python
merged.to_csv(directory + '울산_전처리_ver4.csv',encoding = 'utf-8-sig', index = False)
```
