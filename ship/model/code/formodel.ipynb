## 선박길이,톤수, 선박 종류 merge

```python
cols = ['선박국적코드','총 톤 수','국제 총톤수', '순 톤수', '내외항구분코드',
       '내외항구분명', '청코드', '청코드명', '선적항 항구 코드', '선적항 항구 코드명', '선적항 국가 코드',
       '선적항 국가 코드명','선박 총길이', '선박 너비','선박 번호', '선박 종류', '선박 종류명','호출부호']
```

```python
dfs = dfs[cols]
```

```python
merged = pd.merge(df,dfs,on = '호출부호',how = 'left')
```
